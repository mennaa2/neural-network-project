##IMPORT LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
```

##LOAD DATASET

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

##DATA PREPROCESSING

```python
# Flatten images (28x28 → 784)
X_train = X_train.reshape(-1, 28*28)
X_test = X_test.reshape(-1, 28*28)

# Normalize (0 → 1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
print(y_train.shape)
```

Handling Missing Values

```python
print("Missing values in X_train:", np.isnan(X_train).sum())
print("Missing values in X_test:", np.isnan(X_test).sum())
print("Missing values in y_train:", np.isnan(y_train).sum())
print("Missing values in y_test:", np.isnan(y_test).sum())
```

##BUILD MODEL FUNCTION (MLP)

```python
def build_model(neurons=128, activation='relu'):
    model = Sequential()

    # Input + Hidden Layer
    model.add(Dense(neurons, activation=activation, input_shape=(784,)))

    # Hidden Layer
    model.add(Dense(64, activation=activation))

    # Dropout (regularization)
    model.add(Dropout(0.3))

    # Output Layer
    model.add(Dense(10, activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

##EXPERIMENT 1 (ReLU)

```python
print(" Experiment 1: ReLU ")

model1 = build_model(neurons=128, activation='relu')

history1 = model1.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

loss1, acc1 = model1.evaluate(X_test, y_test)

print("Experiment 1 Accuracy:", acc1)
print("Experiment 1 Loss:", loss1)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = model1.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true, y_pred_classes)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix - MNIST")
plt.show()
```

##EXPERIMENT 2 (Tanh)

```python
print("Experiment 2: Tanh")

model2 = build_model(neurons=128, activation='tanh')

history2 = model2.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

loss2, acc2 = model2.evaluate(X_test, y_test)

print("Experiment 2 Accuracy:", acc2)
print("Experiment 2 Loss:", loss2)
```

```python
y_pred2 = model2.predict(X_test)
y_pred2_classes = np.argmax(y_pred2, axis=1)

cm2 = confusion_matrix(y_true, y_pred2_classes)

ConfusionMatrixDisplay(cm2).plot(cmap='Blues')
plt.title("Confusion Matrix - Exp 2")
plt.show()
```

#EXPERIMENT 3 (ReLU with more neurons)

```python
model_exp3 = build_model(activation='relu', neurons=256)

history_exp3 = model_exp3.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=20,
    batch_size=32,
    verbose=1
)

loss_exp3, acc_exp3 = model_exp3.evaluate(X_test, y_test)
print("Experiment 3 Accuracy:", acc_exp3)
print("Experiment 3 Loss:", loss_exp3)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred3 = model_exp3.predict(X_test)
y_pred3_classes = np.argmax(y_pred3, axis=1)

y_true = np.argmax(y_test, axis=1)
cm3 = confusion_matrix(y_true, y_pred3_classes)

ConfusionMatrixDisplay(cm3).plot(cmap='Blues')
plt.title("Confusion Matrix - Exp 3")
plt.show()
```

##RESULTS COMPARISON

```python
print(" FINAL COMPARISON")

print("{:<10} {:<15} {:<10} {:<10}".format("Model", "Activation", "Accuracy", "Loss"))

print("{:<10} {:<15} {:<10.4f} {:<10.4f}".format("Model 1", "ReLU", acc1, loss1))
print("{:<10} {:<15} {:<10.4f} {:<10.4f}".format("Model 2", "Tanh", acc2, loss2))
print("{:<10} {:<15} {:<10.4f} {:<10.4f}".format("Model 1", "ReLU 256", acc_exp3, loss_exp3))
```

# Handwritten Digit Recognition using MLP (MNIST)

## 📌 Project Overview
This project focuses on building a Multilayer Perceptron (MLP) neural network to classify handwritten digits (0–9) using the MNIST dataset.

The goal is to train a model that can recognize digits from grayscale images with high accuracy.

---

## 📊 Dataset Description
The MNIST dataset contains:

- 60,000 training images  
- 10,000 testing images  
- Each image is 28×28 pixels (grayscale)  
- Labels range from 0 to 9  

Each image represents a handwritten digit.

Dataset Link: https://keras.io/api/datasets/mnist/

---

## ⚙️ Data Preprocessing
The following preprocessing steps were applied:

- Flatten images from 28×28 → 784 vector  
- Normalize pixel values to range [0, 1]  
- Apply one-hot encoding to labels  

These steps help improve model performance and stability.

The dataset does not contain missing values.

---

## 🧠 Model Architecture (MLP)
The model consists of:

- Input Layer: 784 neurons  
- Hidden Layer 1: 128 neurons (ReLU / Tanh)  
- Hidden Layer 2: 64 neurons  
- Dropout Layer: 0.3  
- Output Layer: 10 neurons (Softmax)

The dataset was split into:

- Training set: 80%  
- Validation set: 20% (via validation_split)  
- Test set: provided separately  

---

## ⚙️ Training Configuration
- Optimizer: Adam  
- Loss Function: Categorical Crossentropy  
- Metrics: Accuracy  
- Epochs: 10–20 (depending on experiment)  
- Batch Size: 32  
- Validation Split: 20%

---

## 🧪 Experiments
Three experiments were conducted by varying activation functions and model capacity:

| Experiment | Activation | Neurons | Accuracy | Loss |
|-----------|------------|----------|----------|------|
| Exp 1     | ReLU       | 128      | 97.38%   | 0.10 |
| Exp 2     | Tanh       | 128      | 97.22%   | 0.09 |
| Exp 3     | ReLU       | 256      | 97.98%      | 0.1061   |

The goal is to analyze the effect of activation functions and model complexity.

---

## 📈 Results and Analysis
- ReLU performed slightly better than Tanh in most cases  
- Increasing neurons improved learning capacity in some cases  
- Dropout helped reduce overfitting and improve generalization  

---

## 🧠 Regularization and Enhancement Techniques
Dropout was applied (rate = 0.3) to reduce overfitting and improve generalization.

Batch Normalization and Data Augmentation were not used because:
- The MNIST dataset is already clean and normalized  
- The model achieved high performance without additional techniques  

---

## 📊 Confusion Matrix (Evaluation Insight)
A confusion matrix was used to evaluate detailed classification performance.

- Diagonal values represent correct predictions  
- Off-diagonal values represent misclassifications  

It helps identify which digits are most frequently confused by the model.

---

## 📊 Evaluation Metrics
- Accuracy: measures correct predictions  
- Loss: measures prediction error  
- Confusion Matrix: provides detailed class-level performance

## 📌 Conclusion
The MLP model successfully classified handwritten digits with high accuracy.

The experiments show that:
- Activation functions affect performance  
- Model size impacts learning capacity  
- Regularization (Dropout) improves generalization  

Overall, ReLU-based models performed better in most cases compared to Tanh.

## Analysis and Discussion

**Why ReLU outperforms Tanh:**
ReLU does not saturate for positive values, which means gradients flow more efficiently during backpropagation. Tanh saturates at both ends (output between -1 and 1), causing vanishing gradients in deeper networks, which slows learning and reduces final accuracy.

**Why 256 neurons (Exp 3) achieved higher accuracy:**
Increasing neurons from 128 to 256 gives the model more capacity to learn complex patterns in the data. With more parameters, the first hidden layer can extract richer features from the 784 input pixels, leading to better generalization on the test set.

**Why Exp 3 also trained for 20 epochs instead of 10:**
More neurons need more training steps to converge. The additional epochs allowed the larger model to fully utilize its capacity without underfitting.

**Overall conclusion:**
The best configuration was Experiment 3 (ReLU, 256 neurons, 20 epochs) with 97.98% accuracy. The combination of a non-saturating activation function and higher model capacity were the key factors in achieving the best performance.
